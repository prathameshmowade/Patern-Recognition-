```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Marks': [75, 82, 90, 65, 88]
}

df = pd.DataFrame(data)

print(df)
```

```python
print(df.head())
print(df.describe())
```

```python
plt.bar(df['Student'], df['Marks'])
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.plot(df['Student'], df['Marks'])
plt.title("Student Marks Trend")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
plt.pie(df['Marks'], labels=df['Student'], autopct='%1.1f%%')
plt.title("Marks Distribution")
plt.show()
```

```python
from sklearn.datasets import load_iris

iris = load_iris()

print(iris.data[:5])
print(iris.target[:5])
```

```python
iris_df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

print(iris_df.head())
```

```python
plt.scatter(
    iris_df['sepal length (cm)'],
    iris_df['sepal width (cm)']
)

plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Iris Dataset")
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import euclidean, cityblock
from nltk.metrics.distance import edit_distance
```

```python
df = pd.read_excel('Online Retail.xlsx')

print("Customer Dataset")
display(df)
```

```python
features = ['Quantity', 'UnitPrice']

X = df[features].dropna()

print("Feature Vectors:")
print(X)
```

```python
cust1 = X.iloc[0].values
cust2 = X.iloc[1].values

print("Customer 1:", cust1)
print("Customer 2:", cust2)
```

```python
eu_dist = euclidean(cust1, cust2)

print("Euclidean Distance =", eu_dist)
```

```python
man_dist = cityblock(cust1, cust2)

print("Manhattan Distance =", man_dist)
```

```python
cos_sim = cosine_similarity([cust1], [cust2])

print("Cosine Similarity =", cos_sim[0][0])
```

```python
name1 = df['Description'][0]
name2 = df['Description'][1]

ed_dist = edit_distance(name1, name2)

print("Edit Distance =", ed_dist)
```

```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Ensure df is defined
df = pd.read_excel('Online Retail.xlsx')

# Drop rows with NaN CustomerID and convert to int for cleaner indexing
df_cleaned = df.dropna(subset=['CustomerID']).copy() # Explicitly make a copy
df_cleaned['CustomerID'] = df_cleaned['CustomerID'].astype(int)

# Ensure features and X are defined, derived from the cleaned DataFrame
features = ['Quantity', 'UnitPrice']
X = df_cleaned[features].dropna()

print(f"Shape of X before sampling: {X.shape}")

# Sample X to a manageable size for demonstration due to potential memory issues
# A 500k x 500k matrix is too large for typical environments.
# Limiting to the first 1000 rows for example.
X_sample = X.head(1000)

print(f"Shape of X used for similarity matrix (sampled): {X_sample.shape}")

cos_matrix = cosine_similarity(X_sample)

similarity_df = pd.DataFrame(
    cos_matrix,
    index=df_cleaned.loc[X_sample.index, 'CustomerID'],
    columns=df_cleaned.loc[X_sample.index, 'CustomerID']
)

print("Customer Similarity Matrix (Sampled)")
print(f"Shape of similarity_df: {similarity_df.shape}")
display(similarity_df)
```
